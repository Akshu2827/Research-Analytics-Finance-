```python
"""
Quantitative Analysis of Maritime Defense Companies Around Chokepoint Disruptions
with Additional Financial Health Metrics (Sharpe, Volatility, M-Score, Z-Score, Outliers)
and Event-Wise CAR Correlation Matrices
"""

import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta
from scipy import stats

# ==========================
# 1. Configuration
# ==========================

COMPANIES = {
    "Lockheed Martin": "LMT",
    "Raytheon Technologies": "RTX",
    "General Dynamics": "GD",
    "Northrop Grumman": "NOC",
    "BAE Systems": "BA.L",
    "Rolls-Royce Holdings": "RR.L",
    "Thales Group (UK Div)": "HO.PA",
    "Mitsubishi Heavy Ind": "7011.T",
    "Kawasaki Heavy Ind": "7012.T",
    "Mitsubishi Electric": "6503.T",
    "Garden Reach Shipbuilders": "GRSE.NS",
    "Mazagon Dock Shipbuilders": "MAZDOCK.NS",
    "Cochin Shipyard": "COCHINSHIP.NS",
}

BENCHMARKS = {
    "US": "^GSPC",
    "UK": "^FTSE",
    "Japan": "^N225",
    "India": "^NSEI",
}

def get_benchmark_for_company(company_ticker):
    if company_ticker in ["LMT", "RTX", "GD", "NOC"]:
        return BENCHMARKS["US"]
    elif company_ticker in ["BA.L", "RR.L", "HO.PA"]:
        return BENCHMARKS["UK"]
    elif company_ticker in ["7011.T", "7012.T", "6503.T"]:
        return BENCHMARKS["Japan"]
    elif company_ticker in ["GRSE.NS", "MAZDOCK.NS", "COCHINSHIP.NS"]:
        return BENCHMARKS["India"]
    else:
        raise ValueError(f"Unknown region for ticker {company_ticker}")

EVENTS = {
    "Strait of Hormuz Blockade (hypothetical)": "2026-04-12",
    "Red Sea Houthi attacks (start)": "2023-12-01",
    "Suez Canal blocked (Ever Given)": "2021-03-23",
    "Hormuz tanker attacks (escalation)": "2019-05-01",
    "Maersk Tigris seizure": "2015-04-28",
    "Tanker War (Iran-Iraq) peak": "1987-08-01",
    "Six-Day War / Suez closure": "1967-06-05",
}

WINDOW_DAYS = 30

# ==========================
# 2. Event study functions
# ==========================

def fetch_data(tickers, start_date, end_date):
    print(f"Downloading data from {start_date} to {end_date} ...")
    data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker', auto_adjust=True)
    if len(tickers) == 1:
        data = pd.DataFrame({tickers[0]: data['Close']})
    else:
        close_data = {}
        for t in tickers:
            if t in data.columns.levels[0]:
                close_data[t] = data[t]['Close']
            else:
                print(f"Warning: {t} not found. Skipping.")
        data = pd.DataFrame(close_data)
    return data

def compute_abnormal_returns(company_returns, benchmark_returns):
    return company_returns - benchmark_returns

def cumulative_abnormal_returns(abnormal_returns, event_date, window_days):
    start = event_date - timedelta(days=window_days)
    end = event_date + timedelta(days=window_days)
    ar_window = abnormal_returns.loc[start:end]
    car = ar_window.cumsum()
    car.index = (car.index - event_date).days
    return car

def run_event_study_for_company(company_name, ticker, event_dates, window_days):
    print(f"\n--- Processing {company_name} ({ticker}) ---")
    benchmark_ticker = get_benchmark_for_company(ticker)
    all_event_dates = [pd.to_datetime(d) for d in event_dates.values()]
    min_date = min(all_event_dates) - timedelta(days=window_days+10)
    max_date = max(all_event_dates) + timedelta(days=window_days+10)
    data = fetch_data([ticker, benchmark_ticker], min_date.strftime("%Y-%m-%d"), max_date.strftime("%Y-%m-%d"))
    if ticker not in data.columns or benchmark_ticker not in data.columns:
        print(f"Missing data for {ticker} or {benchmark_ticker}. Skipping.")
        return None, None
    company_returns = data[ticker].pct_change().dropna()
    benchmark_returns = data[benchmark_ticker].pct_change().dropna()
    common_idx = company_returns.index.intersection(benchmark_returns.index)
    company_returns = company_returns[common_idx]
    benchmark_returns = benchmark_returns[common_idx]
    abnormal_ret = compute_abnormal_returns(company_returns, benchmark_returns)
    car_results = {}
    for event_name, event_date_str in event_dates.items():
        event_date = pd.to_datetime(event_date_str)
        if event_date < common_idx.min() or event_date > common_idx.max():
            print(f"Event '{event_name}' on {event_date_str} outside data range. Skipping.")
            continue
        car = cumulative_abnormal_returns(abnormal_ret, event_date, window_days)
        car_results[event_name] = car
    if car_results:
        car_df = pd.DataFrame(car_results)
        car_df.index.name = "Days from event"
        return car_df, company_returns
    else:
        return None, company_returns

# ==========================
# 3. Sharpe ratio & volatility
# ==========================

def compute_sharpe_volatility(returns, risk_free_rate=0.02):
    if len(returns) < 2:
        return np.nan, np.nan
    daily_rf = (1 + risk_free_rate) ** (1/252) - 1
    excess_returns = returns - daily_rf
    sharpe_daily = excess_returns.mean() / excess_returns.std() if excess_returns.std() != 0 else np.nan
    sharpe_annual = sharpe_daily * np.sqrt(252)
    vol_annual = returns.std() * np.sqrt(252)
    return sharpe_annual, vol_annual

# ==========================
# 4. Beneish M-Score
# ==========================

def compute_beneish_m_score(ticker):
    try:
        stock = yf.Ticker(ticker)
        bs = stock.balance_sheet
        is_ = stock.financials
        if bs.empty or is_.empty:
            return np.nan, np.nan
        bs = bs.iloc[:, 0]
        is_ = is_.iloc[:, 0]
        
        def get_value(df, possible_names):
            for name in possible_names:
                if name in df.index:
                    return df[name]
            return np.nan
        
        receivables = get_value(bs, ['Net Receivables', 'Accounts Receivable', 'Trade Receivables'])
        sales = get_value(is_, ['Total Revenue', 'Revenue'])
        dsri = 1.0  # placeholder
        
        cogs = get_value(is_, ['Cost Of Revenue', 'Cost of Goods Sold'])
        if pd.notna(sales) and pd.notna(cogs):
            gross_margin = (sales - cogs) / sales
            gmi = 1.0 / gross_margin if gross_margin > 0 else 1.0
        else:
            gmi = 1.0
        
        total_assets = get_value(bs, ['Total Assets'])
        current_assets = get_value(bs, ['Total Current Assets'])
        if pd.notna(total_assets) and pd.notna(current_assets):
            non_current_assets = total_assets - current_assets
            aqi = non_current_assets / total_assets
        else:
            aqi = 1.0
        
        sgi = 1.0
        depi = 1.0
        sgai = 1.0
        
        total_liab = get_value(bs, ['Total Liabilities Net Minority Interest', 'Total Liabilities'])
        if pd.notna(total_assets) and pd.notna(total_liab):
            leverage = total_liab / total_assets
            lvgi = leverage
        else:
            lvgi = 1.0
        
        net_income = get_value(is_, ['Net Income', 'Net Income Common Stockholders'])
        cash_from_ops = get_value(is_, ['Operating Cash Flow'])
        if pd.notna(net_income) and pd.notna(cash_from_ops) and pd.notna(total_assets):
            tata = (net_income - cash_from_ops) / total_assets
        else:
            tata = 0.0
        
        m_score = (-4.84) + 0.92*dsri + 0.528*gmi + 0.404*aqi + 0.892*sgi + 0.115*depi - 0.172*sgai + 4.679*tata - 0.327*lvgi
        prob = 1 / (1 + np.exp(-m_score))
        return m_score, prob
    except Exception as e:
        print(f"Error computing M-Score for {ticker}: {e}")
        return np.nan, np.nan

# ==========================
# 5. Altman Z-Score
# ==========================

def compute_altman_z_score(ticker, is_emerging_market=False):
    try:
        stock = yf.Ticker(ticker)
        bs = stock.balance_sheet
        is_ = stock.financials
        if bs.empty or is_.empty:
            return np.nan
        bs = bs.iloc[:, 0]
        is_ = is_.iloc[:, 0]
        
        def get_value(df, possible_names):
            for name in possible_names:
                if name in df.index:
                    return df[name]
            return np.nan
        
        current_assets = get_value(bs, ['Total Current Assets', 'Current Assets'])
        current_liab = get_value(bs, ['Total Current Liabilities', 'Current Liabilities'])
        working_capital = (current_assets - current_liab) if pd.notna(current_assets) and pd.notna(current_liab) else np.nan
        total_assets = get_value(bs, ['Total Assets'])
        retained_earnings = get_value(bs, ['Retained Earnings'])
        ebit = get_value(is_, ['EBIT', 'Operating Income'])
        
        market_cap = stock.info.get('marketCap', np.nan)
        if pd.isna(market_cap) and ticker.endswith('.NS'):
            shares = stock.info.get('sharesOutstanding', np.nan)
            price = stock.info.get('regularMarketPrice', np.nan)
            if pd.notna(shares) and pd.notna(price):
                market_cap = shares * price
        
        total_liab = get_value(bs, ['Total Liabilities Net Minority Interest', 'Total Liabilities'])
        sales = get_value(is_, ['Total Revenue', 'Revenue'])
        
        X1 = working_capital / total_assets if pd.notna(working_capital) and pd.notna(total_assets) and total_assets != 0 else np.nan
        X2 = retained_earnings / total_assets if pd.notna(retained_earnings) and pd.notna(total_assets) and total_assets != 0 else np.nan
        X3 = ebit / total_assets if pd.notna(ebit) and pd.notna(total_assets) and total_assets != 0 else np.nan
        X4 = market_cap / total_liab if pd.notna(market_cap) and pd.notna(total_liab) and total_liab != 0 else np.nan
        X5 = sales / total_assets if pd.notna(sales) and pd.notna(total_assets) and total_assets != 0 else np.nan
        
        if is_emerging_market:
            book_equity = get_value(bs, ['Total Equity Gross Minority Interest', 'Stockholders Equity'])
            X4_prime = book_equity / total_liab if pd.notna(book_equity) and pd.notna(total_liab) and total_liab != 0 else np.nan
            z_score = 6.56 * X1 + 3.26 * X2 + 6.72 * X3 + 1.05 * X4_prime
        else:
            z_score = 1.2 * X1 + 1.4 * X2 + 3.3 * X3 + 0.6 * X4 + 1.0 * X5
        return z_score
    except Exception as e:
        print(f"Error computing Z-Score for {ticker}: {e}")
        return np.nan

# ==========================
# 6. Outlier detection
# ==========================

def get_financial_ratios(ticker):
    try:
        stock = yf.Ticker(ticker)
        bs = stock.balance_sheet
        is_ = stock.financials
        if bs.empty or is_.empty:
            return {}
        bs = bs.iloc[:, 0]
        is_ = is_.iloc[:, 0]
        
        def get_value(df, possible_names):
            for name in possible_names:
                if name in df.index:
                    return df[name]
            return np.nan
        
        total_assets = get_value(bs, ['Total Assets'])
        total_liab = get_value(bs, ['Total Liabilities Net Minority Interest', 'Total Liabilities'])
        equity = get_value(bs, ['Total Equity Gross Minority Interest', 'Stockholders Equity'])
        revenue = get_value(is_, ['Total Revenue', 'Revenue'])
        net_income = get_value(is_, ['Net Income', 'Net Income Common Stockholders'])
        current_assets = get_value(bs, ['Total Current Assets', 'Current Assets'])
        current_liab = get_value(bs, ['Total Current Liabilities', 'Current Liabilities'])
        
        ratios = {}
        if pd.notna(current_assets) and pd.notna(current_liab) and current_liab != 0:
            ratios['Current Ratio'] = current_assets / current_liab
        if pd.notna(total_liab) and pd.notna(equity) and equity != 0:
            ratios['Debt/Equity'] = total_liab / equity
        if pd.notna(net_income) and pd.notna(equity) and equity != 0:
            ratios['ROE'] = net_income / equity
        if pd.notna(net_income) and pd.notna(revenue) and revenue != 0:
            ratios['Net Margin'] = net_income / revenue
        if pd.notna(total_liab) and pd.notna(total_assets) and total_assets != 0:
            ratios['Debt/Assets'] = total_liab / total_assets
        return ratios
    except Exception as e:
        print(f"Error computing ratios for {ticker}: {e}")
        return {}

def detect_outliers(all_ratios_dict):
    outliers = {}
    if not all_ratios_dict:
        return outliers
    ratio_names = set()
    for ratios in all_ratios_dict.values():
        ratio_names.update(ratios.keys())
    for ratio in ratio_names:
        values = []
        companies = []
        for comp, rat in all_ratios_dict.items():
            if ratio in rat and pd.notna(rat[ratio]):
                values.append(rat[ratio])
                companies.append(comp)
        if len(values) < 2:
            continue
        z_scores = stats.zscore(values)
        for comp, z in zip(companies, z_scores):
            if abs(z) > 3:
                if comp not in outliers:
                    outliers[comp] = []
                outliers[comp].append((ratio, z))
    return outliers

# ==========================
# 7. Plotting functions
# ==========================

def plot_cumulative_abnormal_returns(car_dict, title="Cumulative Abnormal Returns"):
    all_event_names = set()
    for car_df in car_dict.values():
        if car_df is not None:
            all_event_names.update(car_df.columns)
    if not all_event_names:
        print("No CAR data to plot.")
        return
    n_events = len(all_event_names)
    fig, axes = plt.subplots(n_events, 1, figsize=(12, 4*n_events), sharex=True)
    if n_events == 1:
        axes = [axes]
    for ax, event_name in zip(axes, sorted(all_event_names)):
        ax.axhline(0, color='black', linestyle='--', linewidth=0.8)
        ax.axvline(0, color='red', linestyle='--', label='Event day')
        for company_name, car_df in car_dict.items():
            if car_df is not None and event_name in car_df.columns:
                ax.plot(car_df.index, car_df[event_name], label=company_name, linewidth=1.5)
        ax.set_title(f"Event: {event_name}")
        ax.set_ylabel("Cumulative Abnormal Return")
        ax.legend(loc='best', fontsize=8)
        ax.grid(True, alpha=0.3)
    axes[-1].set_xlabel("Trading days relative to event")
    plt.suptitle(title, fontsize=14, y=1.02)
    plt.tight_layout()
    plt.show()

def plot_individual_company_car(company_name, car_df):
    if car_df is None or car_df.empty:
        print(f"No CAR data for {company_name}")
        return
    plt.figure(figsize=(12, 6))
    for event_name in car_df.columns:
        plt.plot(car_df.index, car_df[event_name], label=event_name, linewidth=1.5)
    plt.axhline(0, color='black', linestyle='--')
    plt.axvline(0, color='red', linestyle='--', label='Event day')
    plt.title(f"Cumulative Abnormal Returns – {company_name}")
    plt.xlabel("Trading days from event")
    plt.ylabel("CAR")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# ==========================
# 8. Correlation helper
# ==========================

def compute_event_correlations(all_car):
    """
    all_car: dict {company_name: DataFrame with columns = events, index = day offsets}
    For each event, collect CAR series for all companies and compute pairwise correlations.
    Returns dict event_name -> correlation DataFrame.
    """
    # First, invert the structure: event -> {company -> Series of CAR over days}
    event_to_companies = {}
    for company, car_df in all_car.items():
        for event_name in car_df.columns:
            if event_name not in event_to_companies:
                event_to_companies[event_name] = {}
            event_to_companies[event_name][company] = car_df[event_name]
    
    correlations = {}
    for event_name, company_series_dict in event_to_companies.items():
        # Create a DataFrame where columns are companies, rows are day offsets
        # Ensure all series have the same index (day offsets). They should because window is fixed.
        common_index = None
        for ser in company_series_dict.values():
            if common_index is None:
                common_index = ser.index
            else:
                # Align indices (should be identical but just in case)
                common_index = common_index.intersection(ser.index)
        if len(common_index) < 2:
            print(f"Event {event_name}: Not enough days to compute correlation.")
            continue
        # Build DataFrame
        df = pd.DataFrame({company: ser.loc[common_index] for company, ser in company_series_dict.items()})
        # Compute correlation matrix
        corr_mat = df.corr()
        correlations[event_name] = corr_mat
    return correlations

# ==========================
# 9. Main execution
# ==========================

if __name__ == "__main__":
    sns.set_style("whitegrid")
    
    all_car = {}
    all_returns = {}
    for company_name, ticker in COMPANIES.items():
        car_df, ret_series = run_event_study_for_company(company_name, ticker, EVENTS, WINDOW_DAYS)
        if car_df is not None:
            all_car[company_name] = car_df
        if ret_series is not None and len(ret_series) > 0:
            all_returns[company_name] = ret_series
    
    if all_car:
        plot_cumulative_abnormal_returns(all_car, title="Defense Companies Around Maritime Chokepoint Events")
        for company_name, car_df in all_car.items():
            plot_individual_company_car(company_name, car_df)
    else:
        print("No valid CAR data generated.")
    
    # Sharpe & Volatility
    sharpe_vol = {}
    for company_name, returns in all_returns.items():
        sharpe, vol = compute_sharpe_volatility(returns)
        sharpe_vol[company_name] = (sharpe, vol)
    
    # Beneish & Altman
    beneish_results = {}
    altman_results = {}
    for company_name, ticker in COMPANIES.items():
        m_score, prob = compute_beneish_m_score(ticker)
        beneish_results[company_name] = (m_score, prob)
        is_india = ticker.endswith('.NS')
        z = compute_altman_z_score(ticker, is_emerging_market=is_india)
        altman_results[company_name] = z
    
    # Outliers
    all_ratios = {}
    for company_name, ticker in COMPANIES.items():
        ratios = get_financial_ratios(ticker)
        all_ratios[company_name] = ratios
    outlier_flags = detect_outliers(all_ratios)
    
    # Print summary
    print("\n" + "="*80)
    print("COMPREHENSIVE FINANCIAL & EVENT STUDY SUMMARY")
    print("="*80)
    print(f"\n{'Company':<35} {'Sharpe':>8} {'Volatility':>10} {'M-Score':>10} {'Prob. Manip':>12} {'Altman Z':>10}")
    print("-"*90)
    for company_name in COMPANIES.keys():
        sharpe, vol = sharpe_vol.get(company_name, (np.nan, np.nan))
        m_score, prob = beneish_results.get(company_name, (np.nan, np.nan))
        z = altman_results.get(company_name, np.nan)
        print(f"{company_name:<35} {sharpe:>8.2f} {vol:>10.2%} {m_score:>10.2f} {prob:>12.2f} {z:>10.2f}")
    
    print("\n" + "-"*80)
    print("FINANCIAL RATIO OUTLIERS (|z-score| > 3)")
    if outlier_flags:
        for company, flags in outlier_flags.items():
            print(f"\n{company}:")
            for ratio, z in flags:
                print(f"  - {ratio}: z = {z:.2f}")
    else:
        print("No extreme outliers detected.")
    
    # Event study CAR max/min
    print("\n" + "-"*80)
    print("EVENT STUDY CUMULATIVE ABNORMAL RETURNS (CAR) – BREAKDOWN BY EVENT")
    for company_name, car_df in all_car.items():
        print(f"\n{company_name}:")
        for event_name in car_df.columns:
            max_car = car_df[event_name].max()
            min_car = car_df[event_name].min()
            print(f"  {event_name}: max CAR = {max_car:.2%}, min CAR = {min_car:.2%}")
    
    # ============================================================
    # NEW: Correlation matrices for each event (time frame)
    # ============================================================
    print("\n" + "-"*80)
    print("EVENT-WISE CORRELATION MATRICES OF CUMULATIVE ABNORMAL RETURNS")
    event_corrs = compute_event_correlations(all_car)
    for event_name, corr_mat in event_corrs.items():
        print(f"\nEvent: {event_name}")
        print(corr_mat.round(2))
        # Optional: plot a heatmap for each event
        plt.figure(figsize=(8, 6))
        sns.heatmap(corr_mat, annot=True, cmap='coolwarm', center=0, fmt='.2f')
        plt.title(f"CAR Correlation Matrix - {event_name}")
        plt.tight_layout()
        plt.show()
    
    print("\n" + "="*80)
    print("INTERPRETATION NOTES:")
    print("  - Sharpe ratio > 1 is good; volatility is annualized standard deviation of daily returns.")
    print("  - Beneish M-Score < -2.22 suggests no manipulation; > -2.22 suggests possible manipulation.")
    print("  - Altman Z-Score: > 2.99 safe, 1.81-2.99 grey, < 1.81 distress zone (original).")
    print("  - For emerging markets Z'' > 2.60 safe, 1.10-2.60 grey, < 1.10 distress.")
    print("  - Outliers flagged by z-score > 3 indicate ratios unusually far from the group mean.")
    print("  - Correlation matrices show how similarly companies' abnormal returns behaved around each event.")
```
